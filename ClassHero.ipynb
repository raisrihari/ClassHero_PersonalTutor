```python
import os
import google.generativeai as genai
from IPython.display import display, Markdown
from google.colab import userdata


# Access the API key securely from Colab's secrets manager
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

if GOOGLE_API_KEY:
    genai.configure(api_key=GOOGLE_API_KEY)
    print("Gemini API key loaded and configured successfully!")
else:
    print("ERROR: GEMINI_API_KEY not found in Colab secrets. Please recheck Step 2.")

# Optional: List available models to confirm connection
print("\nAvailable Gemini Models:")
for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        print(f"- {m.name}")

# Select the model we'll be using for text generation
# We'll stick to 'gemini-pro' for general conversational tasks as it's balanced and powerful.
model = genai.GenerativeModel('gemini-2.5-flash-lite')
print(f"\nSelected model: {model.model_name}")
```

```python
# Create a conversational object (like a chat session)
chat = model.start_chat(history=[])

# Send a message and get a response
prompt = "Hello, Gemini! Can you tell me an interesting fact about artificial intelligence?"
print(f"Me: {prompt}")

response = chat.send_message(prompt)

# Display the response
print(f"Gemini: {response.text}")
```

```python
from google.colab import files
from PyPDF2 import PdfReader # Using PyPDF2 (or pypdf) for extraction

def upload_and_extract_pdf():
    uploaded = files.upload() # This will open a file upload widget

    if not uploaded:
        print("No file uploaded. Please upload a PDF.")
        return None

    # Assuming only one file is uploaded for simplicity
    file_name = next(iter(uploaded))
    print(f"Uploaded file: {file_name}")

    text_content = ""
    try:
        reader = PdfReader(file_name)
        for page in reader.pages:
            text_content += page.extract_text() + "\n" # Add newline for separation
        print(f"Successfully extracted text from '{file_name}'. Total characters: {len(text_content)}")
    except Exception as e:
        print(f"Error processing PDF: {e}")
        return None
    finally:
        # Clean up the uploaded file from the Colab environment after processing
        os.remove(file_name)
        print(f"Removed uploaded file '{file_name}' from Colab runtime.")

    return text_content

# --- Run this function to upload your PDF ---
raw_document_text = upload_and_extract_pdf()

if raw_document_text:
    # Optionally, print the first 500 characters to verify extraction
    print("\n--- First 500 characters of extracted text ---")
    print(raw_document_text[:500])
    print("------------------------------------------")
else:
    print("PDF text extraction failed or no PDF was uploaded.")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

if raw_document_text:
    # Initialize the text splitter
    # character_overlap helps maintain context between chunks
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,      # Max characters per chunk
        chunk_overlap=200,    # Overlap between chunks
        length_function=len,  # How to measure chunk length
        is_separator_regex=False, # Use standard separators
    )

    # Split the document text into chunks
    document_chunks = text_splitter.split_text(raw_document_text)

    print(f"Original text length: {len(raw_document_text)} characters")
    print(f"Number of chunks created: {len(document_chunks)}")
    print("\n--- First 2 chunks ---")
    for i, chunk in enumerate(document_chunks[:2]):
        print(f"Chunk {i+1} (length {len(chunk)}):\n{chunk[:300]}...\n") # Print first 300 chars of chunk
    print("----------------------")
else:
    print("No document text available for chunking. Please re-run Step 6.")
```

```python
from sentence_transformers import SentenceTransformer

# Initialize the embedding model
# 'all-MiniLM-L6-v2' is a good, fast, and relatively small model
# It's suitable for semantic search and works well for many tasks.
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

print(f"Embedding model loaded: {embedding_model}")
print("This model will convert our text chunks into numerical vectors.")
```

```python
import chromadb

if document_chunks:
    persist_directory = "./chroma_db"
    client = chromadb.PersistentClient(path=persist_directory)
    print(f"ChromaDB client initialized, data will be stored in: {persist_directory}")

    collection_name = "pdf_tutor_chunks"

    # --- NEW: Delete existing collection if it exists to prepare for new PDF ---
    try:
        client.delete_collection(name=collection_name)
        print(f"Existing ChromaDB collection '{collection_name}' deleted.")
    except Exception as e:
        # This will gracefully fail if the collection doesn't exist yet (first run)
        print(f"Could not delete collection (might not exist yet): {e}")
    # --- END NEW ---

    # Now, create a fresh collection
    collection = client.get_or_create_collection(name=collection_name)
    print(f"ChromaDB collection '{collection_name}' ready.")

    # Generate embeddings and prepare data for ChromaDB
    chunk_ids = [f"chunk_{i}" for i in range(len(document_chunks))]
    chunk_embeddings = embedding_model.encode(document_chunks).tolist() # Convert to list for ChromaDB

    # Add the chunks to the ChromaDB collection
    print(f"Adding {len(document_chunks)} chunks to ChromaDB...")
    try:
        collection.add(
            embeddings=chunk_embeddings,
            documents=document_chunks,
            ids=chunk_ids
        )
        print(f"Successfully added {len(document_chunks)} chunks to ChromaDB.")
        print(f"Total items in collection: {collection.count()}")

        # Optional: Test a quick query to confirm retrieval works
        test_query = "What is the main topic of this document?"
        results = collection.query(
            query_embeddings=embedding_model.encode([test_query]).tolist(),
            n_results=1
        )
        print("\n--- ChromaDB Test Query Result (most similar chunk) ---")
        print(results['documents'][0][0][:500]) # Print first 500 chars of the top result
        print("--------------------------------------------------")

    except Exception as e:
        print(f"Error adding chunks to ChromaDB: {e}")
        print("Please ensure your embedding model is correctly initialized and document_chunks are available.")

else:
    print("No document chunks available for embedding and database creation. Please re-run Step 7.")
```

```python
# Function to retrieve relevant chunks from ChromaDB
def retrieve_chunks(query, num_results=5):
    """
    Embeds the query and searches ChromaDB for the most relevant text chunks.
    """
    if not collection: # Ensure collection is initialized
        print("Error: ChromaDB collection not initialized.")
        return []

    query_embedding = embedding_model.encode([query]).tolist()
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=num_results,
        include=['documents'] # We only need the text content
    )
    # The documents are nested lists, so flatten them
    retrieved_documents = [doc for sublist in results['documents'] for doc in sublist]
    return retrieved_documents

print("Retrieval function defined.")
```

```python
# Function to generate AI response using Gemini and retrieved chunks
def generate_ai_response(user_question, chat_history):
    """
    Generates a response from Gemini, incorporating retrieved PDF chunks
    and adhering to the 'class topper friend' persona, while also considering
    the conversation history.
    """
    # 1. Retrieve relevant chunks from the PDF based on the user's LATEST question
    # This ensures that even follow-up questions trigger a fresh retrieval
    relevant_chunks = retrieve_chunks(user_question)
    context = "\n".join(relevant_chunks)

    # Convert Gradio history format to Gemini history format
    # Gradio history is list of tuples: [("User message", "Bot response"), ...]
    # Gemini history is list of dicts: [{'role': 'user', 'parts': ['text']}, {'role': 'model', 'parts': ['text']}]
    gemini_history = []
    for human, ai in chat_history:
        gemini_history.append({'role': 'user', 'parts': [human]})
        gemini_history.append({'role': 'model', 'parts': [ai]})

    # 2. Craft a detailed prompt for Gemini, including the retrieved context, persona, and history instruction
    # The crucial part is telling Gemini to consider the *entire conversation* and the *document context*.

    system_instruction = f"""
    You are an AI assistant designed to act as a friendly, clear, and encouraging 'class topper friend'.
    Your goal is to teach the student about topics from the provided document, giving real-world, relatable examples.
    You *must* only answer questions based on the provided document content below.
    If the question cannot be answered from the document, gently explain that you can't find the information in the document and suggest rephrasing or asking about a different topic within the document.
    Do NOT make up information.
    Do NOT go outside the provided context.
    Encourage follow-up questions.
    You can also suggest a small quiz or exercise related to the topic if appropriate.
    Always consider the entire conversation history when formulating your response.

    Document Content (Relevant Sections for current query):
    ---
    {context if context.strip() else "No highly relevant information was found for this specific query within the document."}
    ---
    """

    # We will pass the history directly to model.start_chat or model.generate_content
    # The actual user_question will be the last item in the chat.
    # Gemini's API handles the system instructions somewhat differently,
    # so we'll pass the `system_instruction` as the first message from the user
    # or ensure it's part of the `generate_content` call if we are not using `start_chat` with history.

    # For `start_chat` to maintain history, we should initialize it with the `gemini_history`
    # and then send the `user_question`.

    # If there's no context, guide the response to politely decline
    if not context.strip():
        # If no relevant context is found, guide the AI to say it can't answer from the PDF
        final_prompt_for_gemini = f"""
        Given the previous conversation and the document, you couldn't find relevant information for "{user_question}".
        Please respond by explaining that the information isn't in the document, as per your role instructions.
        """
        # Append this to history for the current turn's generation
        gemini_history.append({'role': 'user', 'parts': [final_prompt_for_gemini]})
    else:
        # Combine the system instruction with the user's current question
        final_prompt_for_gemini = f"""
        {system_instruction}

        Based on the above document content and our ongoing conversation, please answer the user's question.
        User Question: {user_question}

        Your friendly, clear, and document-based explanation (and optional quiz/exercise):
        """
        # Append this to history for the current turn's generation
        gemini_history.append({'role': 'user', 'parts': [final_prompt_for_gemini]})


    # 3. Create a chat session with Gemini and send the prompt
    # Initializing with history now!
    chat_session = model.start_chat(history=gemini_history[:-1]) # Pass all history except the very last user prompt

    try:
        # Send only the latest user prompt, the model already has the history
        response = chat_session.send_message(gemini_history[-1]['parts'][0])

        if response.candidates and response.candidates[0].content.parts:
            return response.candidates[0].content.parts[0].text
        else:
            return "I couldn't generate a coherent response based on our conversation and the document. Please try a different question."
    except Exception as e:
        print(f"Error generating Gemini response: {e}")
        return "Oops! I encountered an error while thinking. Please try again."

print("Response generation function updated for follow-up questions.")
```

```python
import gradio as gr

# Gradio interface function
def chat_with_tutor(message, history):
    """
    Gradio interface wrapper for our AI tutor.
    'history' is a list of [user_message, ai_message] pairs.
    """
    # chat_history from Gradio is a list of lists: [[user1, bot1], [user2, bot2]]
    # Our generate_ai_response currently doesn't directly use this for RAG,
    # but we pass it as a parameter if we wanted to implement full conversational history
    # for the LLM itself in the future.
    response_text = generate_ai_response(message, history)
    return response_text

# Create the Gradio interface
if document_chunks: # Only launch if we have PDF content loaded
    print("Launching Gradio interface...")
    gr.ChatInterface(
        fn=chat_with_tutor,
        title="📚 My Personalized AI Tutor (Class Topper Friend Edition)",
        description="Ask me anything about the uploaded document! I'll explain like your smartest friend, provide examples, and can even suggest quizzes. I'll stick strictly to the document content.",
        examples=[
            ["What is the main topic of this document?"],
            ["Can you explain [a specific term from your PDF]?"],
            ["Give me a small quiz on [another topic from your PDF]."]
        ],
        chatbot=gr.Chatbot(height=500), # Make the chatbot window taller
        theme="soft" # A nice, clean theme
    ).queue().launch(debug=True)
else:
    print("Gradio interface cannot be launched because no PDF content was processed. Please ensure Steps 6-9 were successful.")
    print("If you closed your Colab, you might need to re-run all cells up to Step 9.")
```
